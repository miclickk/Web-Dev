```python
def min4(a, b, c, d):
    m = a
    if b < m:
        m = b
    if c < m:
        m = c
    if d < m:
        m = d
    return m


a, b, c, d = map(int, input().split())

print(min4(a, b, c, d))
```

```python
def power(a, n):
    result = 1
    for i in range(n):
        result *= a
    return result


a, n = map(float, input().split())
n = int(n)

print(power(a, n))
```

```python
def xor(x, y):
    return (x == 1 and y == 0) or (x == 0 and y == 1)


x, y = map(int, input().split())

print(1 if xor(x, y) else 0)
```
