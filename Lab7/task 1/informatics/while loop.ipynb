```python
n = int(input())

i = 1
while i * i <= n:
    print(i * i)
    i += 1
```

```python
n = int(input())

i = 2
while i <= n:
    if n % i == 0:
        print(i)
        break
    i += 1
```

```python
n = int(input())

p = 1

while p <= n:
    print(p, end=" ")
    p = p * 2
```

```python
n = int(input())

while n % 2 == 0 and n > 1:
    n = n // 2

if n == 1:
    print("YES")
else:
    print("NO")
```
